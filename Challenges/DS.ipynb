```python
from sklearn.datasets import load_iris
import pandas as pd


iris = load_iris()

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
iris_df.head()
```

```python

iris_df.info()
iris_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(iris_df, hue='species')
plt.show()
```

```python
iris_df.to_csv('iris_dataset.csv', index=False)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

print("Missing values in Iris dataset:")
print(iris_df.isnull().sum())


iris_df.drop('species', axis=1).hist(figsize=(10, 6), bins=20)
plt.suptitle('Distribution of Features in Iris Dataset')
plt.show()


plt.figure(figsize=(8, 6))
sns.heatmap(iris_df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Features')
plt.show()


sns.pairplot(iris_df, hue='species', palette='Set1')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df.drop('species', axis=1))
plt.title('Boxplot of Iris Dataset Features')
plt.show()


sns.countplot(x='species', data=iris_df)
plt.title('Species Distribution in Iris Dataset')
plt.show()
```

```python

nvidia.isnull().sum()
```

```python

nvidia.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(nvidia['Close'], label='Close Price', color='blue')
plt.title('NVIDIA Stock Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python

nvidia['50_MA'] = nvidia['Close'].rolling(window=50).mean()
nvidia['200_MA'] = nvidia['Close'].rolling(window=200).mean()

plt.figure(figsize=(10, 6))
plt.plot(nvidia['Close'], label='Close Price', color='blue')
plt.plot(nvidia['50_MA'], label='50-Day Moving Average', color='red')
plt.plot(nvidia['200_MA'], label='200-Day Moving Average', color='green')
plt.title('NVIDIA Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python

nvidia['Daily_Return'] = nvidia['Close'].pct_change()


plt.figure(figsize=(10, 6))
plt.hist(nvidia['Daily_Return'].dropna(), bins=50, color='purple', alpha=0.7)
plt.title('NVIDIA Daily Returns Distribution')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=nvidia['Close'], color='orange')
plt.title('Boxplot of NVIDIA Stock Closing Price')
plt.ylabel('Price (USD)')
plt.show()
```

```python

correlation_matrix = nvidia[['Open', 'High', 'Low', 'Close', 'Volume']].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of NVIDIA Stock Features')
plt.show()
```
